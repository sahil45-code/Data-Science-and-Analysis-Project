```python
import pandas as pd

df = pd.read_csv('survey_data_dummy.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Age_Group', palette='Set2')
plt.title("Age Group Distribution")
plt.show()
```

```python
platform_counts = df['Preferred_Platform'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Preferred Streaming Platform")
plt.show()
```

```python
pivot_table = pd.pivot_table(df, values='Satisfaction_Rating',
                              index='Gender', columns='Age_Group', aggfunc='mean')

sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title("Average Satisfaction Rating by Gender & Age Group")
plt.show()
```

```python
import plotly.express as px

fig = px.histogram(df, x="Satisfaction_Rating", color="Gender", barmode="group")
fig.update_layout(title="Satisfaction Rating by Gender")
fig.show()
```
